```python
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
import seaborn as sns
path = '/Users/killaars/PhD/PhD_GIT/mmix_1990120100_1991010100_glb300x200.nc'

fh = nc.Dataset(path,mode='r')

sf6 = fh.variables['nh'][:]

print(np.shape(sf6))

plt.contourf(sf6.mean(axis=2))
plt.show()
plt.contourf(sf6[10,:,:])
plt.show()
```

```python
import datetime as dt

start = dt.datetime(1988,1,1)
end = dt.datetime(1990,12,15,12)

difference = end-start
differenceindays = difference.days+difference.seconds/(3600*24)
differenceindays

forcing = 1e-15*3600*24
forcing

nh_age = differenceindays-sf6/forcing
V=np.arange(-10,1000,20)
plt.contourf(nh_age.mean(axis=2),V)
plt.colorbar()
plt.show()
```
