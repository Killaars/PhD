```python
import numpy as np
import pandas as pd
import os,sys
import netCDF4 as nc
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

path='/Users/killaars/PhD/Elnino/Discharge/grdc/MSWEP_AMA/'

for file in os.listdir(path):
    if file.endswith('txt'):
        print(file)
        data=pd.read_csv(os.path.join(path,file),sep=';',skiprows=3)
        data["new_date"]=pd.to_datetime(data["date"],format='%Y/%m/%d')
        data=data.set_index('new_date')
        data=data.dropna()
        
        scores=pd.read_csv(os.path.join(path,file),sep=';',skiprows=1,nrows=1)
        print('KGE:',scores['KGE_2012'].values,'Correlation:',scores['correlation'].values)
        
        # -- Create figure
        fig = plt.figure(figsize=[30,20])
        gs = GridSpec(1,1,width_ratios=[1],height_ratios=[1.5])
        #gs.update(wspace=0.13,hspace=0.1)
        
        # Ax 3 and 4: Discharge at Obidos and Caracarai
        ax1 = fig.add_subplot(gs[0,0])
        ax1.plot(data.index, data['observation'], 'r',label='Observations')
        ax1.plot(data.index, data['simulation'], 'b',label='PCR-GLOBWB')
        plt.ylabel('Discharge (m3/s)')
        plt.legend(frameon=False)
        plt.grid(True)
        plt.show()
        

sys.exit()
```
